```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("EnterpriseDW.db")

print("Database Created Successfully")
```

```python
date = pd.read_csv("Dim_Date.csv")

customers = pd.read_csv("Dim_Customers.csv")

products = pd.read_csv("Dim_Products.csv")

stores = pd.read_csv("Dim_Stores.csv")

suppliers = pd.read_csv("Dim_Suppliers.csv")

sales = pd.read_csv("Fact_Sales.csv")

inventory = pd.read_csv("Fact_Inventory.csv")

marketing = pd.read_csv("Fact_Marketing.csv")

support = pd.read_csv("Fact_Support.csv")
```

```python
date.to_sql(
    "Dim_Date",
    conn,
    if_exists="replace",
    index=False
)

customers.to_sql(
    "Dim_Customers",
    conn,
    if_exists="replace",
    index=False
)

products.to_sql(
    "Dim_Products",
    conn,
    if_exists="replace",
    index=False
)

stores.to_sql(
    "Dim_Stores",
    conn,
    if_exists="replace",
    index=False
)

suppliers.to_sql(
    "Dim_Suppliers",
    conn,
    if_exists="replace",
    index=False
)

sales.to_sql(
    "Fact_Sales",
    conn,
    if_exists="replace",
    index=False
)

inventory.to_sql(
    "Fact_Inventory",
    conn,
    if_exists="replace",
    index=False
)

marketing.to_sql(
    "Fact_Marketing",
    conn,
    if_exists="replace",
    index=False
)

support.to_sql(
    "Fact_Support",
    conn,
    if_exists="replace",
    index=False
)

print("All Tables Imported Successfully")
```
