```python
import pandas as pd
import numpy as np
from faker import Faker
import random
from datetime import datetime, timedelta
```

```python
fake = Faker('en_IN')

random.seed(42)
np.random.seed(42)

NUM_CUSTOMERS = 10000

CURRENT_DATE = datetime.today()
```

```python
cities = {
    "Mumbai": "Maharashtra",
    "Pune": "Maharashtra",
    "Delhi": "Delhi",
    "Noida": "Uttar Pradesh",
    "Lucknow": "Uttar Pradesh",
    "Jaipur": "Rajasthan",
    "Ahmedabad": "Gujarat",
    "Bengaluru": "Karnataka",
    "Hyderabad": "Telangana",
    "Chennai": "Tamil Nadu",
    "Kolkata": "West Bengal",
    "Indore": "Madhya Pradesh",
    "Chandigarh": "Chandigarh",
    "Kochi": "Kerala",
    "Gurugram": "Haryana"
}

segments = ["Consumer","Corporate","Small Business"]
segment_prob = [0.70,0.20,0.10]

channels = ["Website","Mobile App","Physical Store"]
channel_prob = [0.35,0.45,0.20]

occupations = [
    "Software Engineer",
    "Teacher",
    "Doctor",
    "Business Owner",
    "Student",
    "Government Employee",
    "Consultant",
    "Sales Executive",
    "Marketing Manager",
    "Chartered Accountant",
    "Lawyer",
    "Freelancer"
]
```

```python
customers = []

for cid in range(1, NUM_CUSTOMERS + 1):

    gender = random.choice(["Male","Female"])

    if gender == "Male":
        first_name = fake.first_name_male()
    else:
        first_name = fake.first_name_female()

    last_name = fake.last_name()

    age = random.randint(18,70)

    dob = CURRENT_DATE - timedelta(days=age*365 + random.randint(0,364))

    city = random.choice(list(cities.keys()))
    state = cities[city]

    annual_income = int(np.random.normal(800000,400000))
    annual_income = max(250000,annual_income)

    registration_date = fake.date_between(
        start_date='-5y',
        end_date='today'
    )

    customer_segment = np.random.choice(
        segments,
        p=segment_prob
    )

    preferred_channel = np.random.choice(
        channels,
        p=channel_prob
    )

    occupation = random.choice(occupations)

    total_orders = random.randint(1,80)

    avg_order_value = int(
        annual_income/180 +
        np.random.normal(1000,500)
    )

    avg_order_value = max(500,avg_order_value)

    lifetime_spend = total_orders * avg_order_value

    if lifetime_spend < 30000:
        loyalty = "Bronze"
    elif lifetime_spend < 80000:
        loyalty = "Silver"
    elif lifetime_spend < 180000:
        loyalty = "Gold"
    else:
        loyalty = "Platinum"

    if loyalty=="Bronze":
        churn = random.randint(60,95)
    elif loyalty=="Silver":
        churn = random.randint(35,65)
    elif loyalty=="Gold":
        churn = random.randint(15,40)
    else:
        churn = random.randint(1,20)

    customers.append([
        cid,
        first_name,
        last_name,
        gender,
        age,
        dob.date(),
        city,
        state,
        fake.postcode(),
        "India",
        registration_date,
        customer_segment,
        loyalty,
        preferred_channel,
        annual_income,
        occupation,
        fake.email(),
        fake.phone_number(),
        total_orders,
        avg_order_value,
        lifetime_spend,
        churn,
        random.choice([True,True,True,False])
    ])
```

```python
columns = [
    "customer_id",
    "first_name",
    "last_name",
    "gender",
    "age",
    "date_of_birth",
    "city",
    "state",
    "pincode",
    "country",
    "registration_date",
    "customer_segment",
    "loyalty_tier",
    "preferred_channel",
    "annual_income",
    "occupation",
    "email",
    "phone",
    "total_orders",
    "avg_order_value",
    "lifetime_spend",
    "churn_risk",
    "is_active"
]

dim_customer = pd.DataFrame(customers, columns=columns)

dim_customer.head()
```

```python
import pandas as pd
import numpy as np
import random
from faker import Faker
from datetime import datetime, timedelta
from google.colab import files
```

```python
fake = Faker("en_IN")

random.seed(42)
np.random.seed(42)

NUM_PRODUCTS = 2000
TODAY = datetime.today()
```

```python
category_dict = {
    "Electronics": {
        "subcategories": ["Mobile","Laptop","Tablet","Headphones","Smartwatch"],
        "brands": ["Apple","Samsung","Sony","Dell","HP","Lenovo","Boat","JBL"],
        "cost": (2000,70000),
        "margin": (0.18,0.35),
        "warranty": (12,24)
    },

    "Fashion": {
        "subcategories": ["Men Clothing","Women Clothing","Shoes","Accessories"],
        "brands": ["Nike","Adidas","Puma","Levis","Allen Solly","Van Heusen"],
        "cost": (300,4000),
        "margin": (0.30,0.60),
        "warranty": (0,0)
    },

    "Home & Kitchen": {
        "subcategories": ["Furniture","Kitchen Appliances","Home Decor"],
        "brands": ["Prestige","Philips","Milton","Borosil"],
        "cost": (500,15000),
        "margin": (0.25,0.45),
        "warranty": (6,24)
    },

    "Beauty": {
        "subcategories": ["Skin Care","Hair Care","Makeup"],
        "brands": ["Lakme","Loreal","Nivea","Maybelline","Dove"],
        "cost": (100,1500),
        "margin": (0.35,0.60),
        "warranty": (0,0)
    },

    "Sports": {
        "subcategories": ["Fitness","Outdoor","Cycling"],
        "brands": ["Nike","Adidas","Decathlon","Puma"],
        "cost": (500,12000),
        "margin": (0.25,0.50),
        "warranty": (6,18)
    },

    "Grocery": {
        "subcategories": ["Snacks","Beverages","Dairy"],
        "brands": ["Amul","Nestle","Britannia","Pepsi","Coca Cola"],
        "cost": (20,600),
        "margin": (0.15,0.30),
        "warranty": (0,0)
    }
}
```

```python
product_names = {

"Mobile":["Ultra","Pro","Max","Lite","Prime","Neo"],

"Laptop":["Book","Air","Notebook","Elite","Slim"],

"Tablet":["Pad","Tab","SmartTab"],

"Headphones":["Bass","Studio","Pods","AirSound"],

"Smartwatch":["Fit","Watch","Pulse"],

"Men Clothing":["Shirt","Jeans","Jacket","T-Shirt"],

"Women Clothing":["Top","Dress","Kurti","Jeans"],

"Shoes":["Running Shoes","Sneakers","Sports Shoes"],

"Accessories":["Wallet","Belt","Cap","Bag"],

"Furniture":["Chair","Table","Sofa","Desk"],

"Kitchen Appliances":["Mixer","Cooker","Toaster"],

"Home Decor":["Lamp","Mirror","Painting"],

"Skin Care":["Face Wash","Moisturizer","Serum"],

"Hair Care":["Shampoo","Conditioner","Hair Oil"],

"Makeup":["Lipstick","Foundation","Mascara"],

"Fitness":["Dumbbell","Yoga Mat","Resistance Band"],

"Outdoor":["Tent","Backpack","Sleeping Bag"],

"Cycling":["Cycle Helmet","Mountain Bike"],

"Snacks":["Chips","Cookies","Namkeen"],

"Beverages":["Juice","Soft Drink","Energy Drink"],

"Dairy":["Milk","Butter","Cheese"]
}
```

```python
products = []

for pid in range(1, NUM_PRODUCTS+1):

    category = random.choice(list(category_dict.keys()))

    info = category_dict[category]

    subcategory = random.choice(info["subcategories"])

    brand = random.choice(info["brands"])

    product_suffix = random.choice(product_names[subcategory])

    product_name = f"{brand} {product_suffix}"

    supplier_id = random.randint(1,300)

    cost_price = random.randint(info["cost"][0],info["cost"][1])

    margin = round(random.uniform(info["margin"][0],info["margin"][1]),2)

    selling_price = round(cost_price*(1+margin),2)

    discount = random.choice([0,5,10,15,20,25])

    launch_date = fake.date_between(start_date="-4y",end_date="today")

    rating = round(random.uniform(3.5,5.0),1)

    review_count = int(abs(np.random.normal(500,250)))

    warranty = random.randint(info["warranty"][0],info["warranty"][1])

    weight = round(random.uniform(0.1,20),2)

    stock_status = np.random.choice(

        ["In Stock","Low Stock","Out of Stock"],

        p=[0.75,0.18,0.07]

    )

    is_active = random.choice([True,True,True,False])

    products.append([

        pid,

        product_name,

        category,

        subcategory,

        brand,

        supplier_id,

        cost_price,

        selling_price,

        discount,

        round(margin*100,2),

        launch_date,

        rating,

        review_count,

        warranty,

        weight,

        stock_status,

        is_active

    ])
```

```python
columns=[

"product_id",

"product_name",

"category",

"sub_category",

"brand",

"supplier_id",

"cost_price",

"selling_price",

"discount_percent",

"profit_margin",

"launch_date",

"product_rating",

"review_count",

"warranty_months",

"weight_kg",

"stock_status",

"is_active"

]

dim_product=pd.DataFrame(products,columns=columns)

dim_product.head()
```

```python
dim_product["discounted_price"] = (
    dim_product["selling_price"] *
    (1-dim_product["discount_percent"]/100)
).round(2)

dim_product["expected_profit"] = (
    dim_product["selling_price"] -
    dim_product["cost_price"]
).round(2)
```

```python
print("="*50)

print("Shape")

print(dim_product.shape)

print("="*50)

print("Null Values")

print(dim_product.isnull().sum())

print("="*50)

print("Duplicate Product IDs")

print(dim_product.product_id.duplicated().sum())

print("="*50)

print("Category Distribution")

print(dim_product.category.value_counts())

print("="*50)

print("Average Selling Price")

print(dim_product.groupby("category")["selling_price"].mean().round(2))

print("="*50)

print("Stock Status")

print(dim_product.stock_status.value_counts())
```

```python
dim_product.to_csv("dim_product.csv",index=False)

print("CSV Saved Successfully")
```

```python
import pandas as pd
import numpy as np
import random

from faker import Faker
from datetime import datetime
from google.colab import files
```

```python
fake = Faker("en_IN")

random.seed(42)
np.random.seed(42)

NUM_STORES = 100
```

```python
city_data = {
    "Mumbai": ("Maharashtra","West",19.0760,72.8777),
    "Pune": ("Maharashtra","West",18.5204,73.8567),
    "Ahmedabad": ("Gujarat","West",23.0225,72.5714),
    "Surat": ("Gujarat","West",21.1702,72.8311),

    "Delhi": ("Delhi","North",28.7041,77.1025),
    "Noida": ("Uttar Pradesh","North",28.5355,77.3910),
    "Lucknow": ("Uttar Pradesh","North",26.8467,80.9462),
    "Chandigarh": ("Chandigarh","North",30.7333,76.7794),
    "Jaipur": ("Rajasthan","North",26.9124,75.7873),

    "Bengaluru": ("Karnataka","South",12.9716,77.5946),
    "Hyderabad": ("Telangana","South",17.3850,78.4867),
    "Chennai": ("Tamil Nadu","South",13.0827,80.2707),
    "Kochi": ("Kerala","South",9.9312,76.2673),
    "Coimbatore": ("Tamil Nadu","South",11.0168,76.9558),

    "Kolkata": ("West Bengal","East",22.5726,88.3639),
    "Bhubaneswar": ("Odisha","East",20.2961,85.8245),
    "Patna": ("Bihar","East",25.5941,85.1376),
    "Guwahati": ("Assam","East",26.1445,91.7362),
    "Ranchi": ("Jharkhand","East",23.3441,85.3096)
}
```

```python
store_types = [
    "Mall",
    "High Street",
    "Express",
    "Warehouse"
]

store_prob = [
    0.30,
    0.30,
    0.25,
    0.15
]
```

```python
stores = []

for sid in range(1, NUM_STORES + 1):

    city = random.choice(list(city_data.keys()))

    state, region, lat, lon = city_data[city]

    store_type = np.random.choice(store_types, p=store_prob)

    if store_type == "Mall":
        size = random.randint(12000,22000)
        employees = random.randint(45,80)
        rent = random.randint(25000000,45000000)

    elif store_type == "High Street":
        size = random.randint(7000,12000)
        employees = random.randint(25,45)
        rent = random.randint(12000000,25000000)

    elif store_type == "Express":
        size = random.randint(2500,6000)
        employees = random.randint(10,25)
        rent = random.randint(4000000,12000000)

    else:
        size = random.randint(15000,30000)
        employees = random.randint(20,45)
        rent = random.randint(15000000,35000000)

    manager = fake.name()

    opening_date = fake.date_between(
        start_date="-15y",
        end_date="today"
    )

    stores.append([
        f"S{sid:03}",
        f"RetailX {city} Store",
        city,
        state,
        region,
        store_type,
        opening_date,
        size,
        employees,
        rent,
        round(lat + random.uniform(-0.03,0.03),6),
        round(lon + random.uniform(-0.03,0.03),6),
        manager,
        random.choice([True,True,True,False])
    ])
```

```python
columns = [

    "store_id",

    "store_name",

    "city",

    "state",

    "region",

    "store_type",

    "opening_date",

    "store_size_sqft",

    "employees",

    "annual_rent",

    "latitude",

    "longitude",

    "manager_name",

    "is_active"

]

dim_store = pd.DataFrame(stores, columns=columns)

dim_store.head()
```

```python
dim_store["rent_per_sqft"] = (
    dim_store["annual_rent"] /
    dim_store["store_size_sqft"]
).round(2)

dim_store["employee_density"] = (
    dim_store["employees"] /
    dim_store["store_size_sqft"] * 1000
).round(2)
```

```python
print("="*50)

print("Shape")

print(dim_store.shape)

print("="*50)

print("Null Values")

print(dim_store.isnull().sum())

print("="*50)

print("Duplicate Store IDs")

print(dim_store.store_id.duplicated().sum())

print("="*50)

print("Region Distribution")

print(dim_store.region.value_counts())

print("="*50)

print("Store Types")

print(dim_store.store_type.value_counts())

print("="*50)

print("Average Employees")

print(
    dim_store.groupby("store_type")["employees"].mean()
)
```
